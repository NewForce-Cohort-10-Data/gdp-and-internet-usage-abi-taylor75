```python
internet_df = internet_df.drop(columns =['Value Footnotes']).rename(columns = {'Country or Area':'Country', 'Value':'Internet_Users_Pct'})
```

```python
internet_counts = internet_df.Year.value_counts().reset_index().sort_values(by = 'Year', ascending=True)
```

```python
internet_counts
```

```python
internet_counts = internet_df[internet_df['Year'].between(1990, 2023)]
```

```python
sns.lineplot(x='Year',y='count', data=internet_counts, marker='o', color='purple')
```

between 1990 and 2000 there was a sharp increase in the number of observations. Between 2000 and 2017 observations were consistent then there is a decrease from 2017 to 2022.

```python
first_year = internet_df.loc[internet_df.Internet_Users_Pct >=1].sort_values(by = 'Year', ascending = True).reset_index(drop=True)
print(first_year.head())
```

```python
internet_2000 = internet_df[internet_df['Year'] == 2000]
internet_2014 = internet_df[internet_df['Year'] == 2014]
```

```python
sns.kdeplot(internet_2000['Internet_Users_Pct'], label='2000', color='blue', clip=(0, 100))
sns.kdeplot(internet_2014['Internet_Users_Pct'], label='2014', color='purple', clip=(0, 100))
plt.xlabel('Internet Users (% of Population)')
plt.ylabel('Density')
plt.title('Distribution of Internet Users Percentage (2000 vs 2014)')
plt.legend()
plt.show()
```

In 2000 there were a lot of countries that had roughly the same amount of users (many being under 20%). Fewer countries had over 20%. In 2014, the distribution of users is more even across all countries. Internet usage across the globe greatly increased over that 14 year period.
